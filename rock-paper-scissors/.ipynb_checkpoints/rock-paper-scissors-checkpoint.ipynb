# Rock Paper Scissors

Let's play the famous game against our computer.

```python
# Import the choice function of the random module
# https://stackoverflow.com/questions/306400/how-to-randomly-select-an-item-from-a-list
from random import choice

# Assign to a list the 3 possible options: 'stone', 'paper' or 'scissors'.
gestures=["rock","paper","scissors"]

# Assign a variable to the maximum number of games: 1, 3, 5, etc ...
n_rounds=5

# Assign a variable to the number of games a player must win to win.
# Preferably the value will be based on the number of maximum games
rounds_to_win=2

# Define a function that randomly returns one of the 3 options.
# This will correspond to the play of the machine. Totally random.
def cpu_choice(gestures):
    cpu_play=choice(gestures)
    return cpu_play

# Define a function that asks your choice: 'stone', 'paper' or 'scissors'
# you should only allow one of the 3 options. This is defensive programming.
# If it is not stone, paper or scissors keep asking until it is.
def input_player(gestures):
    player_play=input("One, Two, Three!! Which choice do you play?"+str(gestures)).lower()
    
    while player_play not in gestures:
            player_play=input("Not valid entry. Repeat again! One, Two, Three!! Which choice do you play? (rock, paper or scissors)")

    return player_play

# Define a function that resolves a combat.
# Returns 0 if there is a tie, 1 if the machine wins, 2 if the human player wins
def check_who_won(player_play,cpu_play):
    if player_play==cpu_play:
        result=0
    elif (cpu_play=="rock" and player_play=="scissors") or (cpu_play=="paper" and player_play=="rock") or (cpu_play=="scissors" and player_play=="paper"):
        result=1
    elif (player_play=="rock" and cpu_play=="scissors") or (player_play=="paper" and cpu_play=="rock") or (player_play=="scissors" and cpu_play=="paper"):
        result=2
        
    return result
    
# Define a function that shows the choice of each player and the state of the game
# This function should be used every time accumulated points are updated
def resolution_play(player_play,cpu_play,result,cpu_score,player_score):
    print("You played "+ player_play+"!!!")
    print("Your opponent played.... "+ cpu_play+"!!!")
    
    if result==0:
        print("Tie")
    elif result==1:
        cpu_score+=1
        print("Round won by CPU")
    elif result==2:
        player_score+=1
        print("Round won by Player")
    return [cpu_score,player_score]
    
# Create two variables that accumulate the wins of each participant
cpu_score=0
player_score=0

# Create a loop that iterates while no player reaches the minimum of wins
# necessary to win. Inside the loop solves the play of the
# machine and ask the player's. Compare them and update the value of the variables
# that accumulate the wins of each participant.
rounds=0
score=[0,0]
cpu_score=0
player_score=0
print("Total rounds: ",n_rounds,". Total wins for victory: ",rounds_to_win)

while (player_score or cpu_score)<int(rounds_to_win) and rounds<int(n_rounds):
    player_play=input_player(gestures)
    cpu_play=cpu_choice(gestures)
    result=check_who_won(player_play,cpu_play)
    score[0]=resolution_play(player_play,cpu_play,result,cpu_score,player_score)
    rounds+=1
    cpu_score=score[0][0]
    player_score=score[0][1]
    
# Print by console the winner of the game based on who has more accumulated wins
if (cpu_score < int(rounds_to_win)) and (player_score<int(rounds_to_win)):
        print("It's a tie! Nobody reached minimum won rounds of"+rounds_to_win)
        print("FYI, Out of "+str(n_rounds)+" rounds, the score was: Player: "+str(player_score)+" won rounds. CPU: "+str(cpu_score)+" won rounds.")
    
elif player_score>cpu_score:
        print("Player wins!! Congratulations")
elif cpu_score>player_score:
        print("CPU player wins!! A machine defeated a real person")
elif player_score==cpu_score:
        print("It's a tie to "+str(player_score)+" points") 
```
